Since we plan to to measure the Hausdorff dimension d, we need to linearize the equation $d^r=N$, where N is the particle number and r is the distance to the biggest possible circle.
Therefore one needs to apply an double logarythmic scale and then fit the function with a linear fit. The gradient is the dimension. Such a linear function is for example the function lin(x,b).

```python
def lin(x,b):
            return 0*x+b
```

Now we need to write the evaluation program. We want that it imports the neded data set and then evaluated it. The following function evaluates it seemingly good.

```python
def evalf(Data,exc,typ,Name,EV,image):
    '''''''''''''''
    
    Data - Import the data set you need to evaluate
    
    exc - Option 1 and 2, both are the same therefore a useless option
    
    typ - options are: raw, ecal,lograw, loglog plot. Each of the plot/present the data differently
    
    Name - gives the figure a name which will be saved
    
    EV - If EV is true a figure will be saved
    
    image - Name of the image wich will be included in the plot
    
    
    '''''''''''''''
    
    Err=[]
    
    if exc==1:
        y=Data[0]*2
        x=Data[1]
        
        #Fit function
        
        cor,cov=np.polyfit(np.log10(x),np.log10(y) ,1, rcond=None, full=False, w=None, cov=True)
        
        logx=np.log(x)
        logy=np.log(y)
        y_fit=cor[0]*logx+cor[1]
        
        Err.append(np.linalg.det(cov))
        
        #What plot do you want?
        
        if typ=='raw':

            #Plot raw 
            plt.figure(figsize=(20,10))
            plt.plot(list(x),list(y),'x')

            plt.figure(figsize=(20,10))
            
        elif typ=='log_raw':


            plt.loglog(x,y,'x')
            
        #Data in normal system
            
        elif typ=='eval':

           

            y_fit=cor[0]*logx+cor[1]

            plt.figure(figsize=(20,10))

            plt.ylabel("$log(N_C)$",fontsize=40)
            plt.xlabel("$log(R)$",fontsize=40)

            plt.yticks(fontsize=25)
            plt.xticks(fontsize=25)

            plt.plot(np.log(x),logy, 'kx', Label='Data Points')
            plt.plot(logx,y_fit,Label='Fit')

            plt.legend(fontsize=30)

        elif typ=='logplot':

            #plt.figure(figsize=(20,10))
            
            fig, ax = plt.subplots(figsize=(10,5))

            ax.set_ylabel("Particle Number $ N_P$",fontsize=30)
            ax.set_xlabel("Cluster Size $R$",fontsize=40)

            #ax.set_xlabel(fontsize=25)
            ax.tick_params(axis='both', which='major', labelsize=22)
            
            #fig, ax = plt.subplots()
            
            ax.loglog(x,y, 'kx', Label='Data Points')
            ax.loglog(x,np.exp(y_fit),Label='Fit')
            #ax.errorbar(y, x, yerr=0.1*np.ones(len(y)),color='k', fmt=' ')
            #plt.errorbar(x,y,yerr=np.ones(len(x)))
            
            print('$r^2$-score is:',r2_score(np.exp(y_fit), y,multioutput='variance_weighted'))
            
            if EV==True:
                
                im = plt.imread('/Users/milos/Desktop/Uni/Master/Semester 2/ComphysB/DLA/Pictures/' + image) # insert local path of the image.
                #fig, ax = plt.subplots()
                #ax.plot(range(100))
                newax = fig.add_axes([0.48,0.15,0.4,0.4], anchor='NE', zorder=4)   ###(lower->left,lower->down,size of image)
                newax.imshow(im)
                newax.axis('off')

                ax.legend(fontsize=20)

                plt.savefig(os.path.join('/Users/milos/Desktop', Name),dpi=None, bbox_inches = "tight"   )
                
            else:
                pass


            
            
        elif typ=='error':
            return np.linalg.det(cov)
            
        else:
            
            #print(cor[0])
            
            return cor[0]
        
    elif exc==2:
        
       
        y=Data[0]
        x=Data[1]
        
        #Fit function
        
        cor,cov=np.polyfit(np.log10(x),np.log10(y) ,1, rcond=None, full=False, w=None, cov=True)
        
        logx=np.log(x)
        logy=np.log(y)
        y_fit=cor[0]*logx+cor[1]
        
        
        
        #What plot do you want?
        
        if typ=='raw':

            #Plot raw 
            plt.figure(figsize=(20,10))
            plt.plot(list(x),list(y),'x')

            plt.figure(figsize=(20,10))
            
        elif typ=='log_raw':


            plt.loglog(x,y,'x')
            
            
        #Data in normal system
            
        elif typ=='eval':

           

            y_fit=cor[0]*logx+cor[1]

            plt.figure(figsize=(20,10))

            plt.ylabel("$log(N_C)$",fontsize=40)
            plt.xlabel("$log(R)$",fontsize=40)

            plt.yticks(fontsize=25)
            plt.xticks(fontsize=25)

            plt.plot(np.log(x),logy, 'kx', Label='Data Points')
            plt.plot(logx,y_fit,Label='Fit')

            plt.legend(fontsize=30)

        elif typ=='logplot':

            plt.figure(figsize=(20,10))

            plt.ylabel("$Log(N_C)$",fontsize=40)
            plt.xlabel("$Log(R)$",fontsize=40)

            plt.yticks(fontsize=25)
            plt.xticks(fontsize=25)

            plt.loglog(x,y, 'kx', Label='Data Points')
            plt.loglog(x,np.exp(y_fit),Label='Fit')

            plt.errorbar(x,y,yerr=np.ones(len(x)))

            plt.legend(fontsize=30)
            
        else:
            
            print(cor[0])
            
            return cor[0]
evalf(Ex1_9000,exc=1,typ='logplot',Name='9000_lolog',EV=True,image='Frac_try.png')
```

looking in at the plot previously (in this notebook a bit later) gives us a shape of somewhat of a powerfunction wich converges to a specific value. Therefore we will fit our data with a fit function of $ f(x)=a\cdot x^{b}+d\cdot x^{e}+c $.
We can use this to estimate the dimension of particle number even much higher than we can simulate with our program, i.e. calculate the limes and see how the function behaves for large numbers.

```python
#definition of our Fit-Function

def power_plot(x,a,b,c,d,e):
    return a*x**b+c+d*x**e

#Fit parameters

popt, pcov = curve_fit(power_plot, Num, dimension_view,maxfev = 15800)
```

Since we have now the image, we can calculate the Heindorf dimensions even with a second method. Antoher well known method is the Box-Counting method.
To use the obtained image for our Box counting algorythm we need to grey scale it first. One such function is given as in the following lines of code.

```python

#Makes Grey scale

def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
 
```

Now we can code the box counting algorythm. 

```python
def Boxcount(im):
    

    
    image=rgb2gray(pl.imread(im))

    # finding all the non-zero pixels
    pixels=[]
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if image[i,j]>0:
                pixels.append((i,j))

    Lx=image.shape[1]
    Ly=image.shape[0]
    print (Lx, Ly)
    pixels=pl.array(pixels)
    print (pixels.shape)

    # computing the fractal dimension
    #considering only scales in a logarithmic list
    scales=np.logspace(0.01, 1, num=10, endpoint=False, base=2)
    Ns=[]
    # looping over several scales
    for scale in scales:
        #print ("======= Scale :",scale)
        # computing the histogram
        H, edges=np.histogramdd(pixels, bins=(np.arange(0,Lx,scale),np.arange(0,Ly,scale)))
        Ns.append(np.sum(H>0))

    # linear fit, polynomial of degree 1
    coeffs=np.polyfit(np.log(scales), np.log(Ns), 1)

    
    plt.figure(figsize=(10,5))
    pl.plot(np.log(scales),np.log(Ns), 'kx', mfc='none')
    pl.plot(np.log(scales), np.polyval(coeffs,np.log(scales)),'r')
    pl.xlabel('Grid Interval log $\omega$',fontsize=30)
    pl.ylabel('Occupied Cells log N',fontsize=30)
    pl.savefig('sierpinski_dimension.pdf')

    print ("The Hausdorff dimension is", -coeffs[0]) #the fractal dimension is the OPPOSITE of the fitting coefficient
    #np.savetxt("scaling.txt", list(zip(scales,Ns)))
```

```python
def variance(data):
...     # Number of observations
...     n = len(data)
...     # Mean of the data
...     mean = sum(data) / n
...     # Square deviations
...     deviations = [(x - mean) ** 2 for x in data]
...     # Variance
...     variance = sum(deviations) / n
...     return variance
```

```python
#Mandelbrot Fractal

from PIL import Image#  eventuell ntig: pip3 install Pillow
import threading

w, h, zoom = 1920, 1080, 1

# creating the new image in RGB mode
bitmap = Image.new("RGB", (w, h), "white")
pix = bitmap.load()

cX, cY = -0.7, 0.27015
moveX, moveY = 0.0, 0.0
maxIter = 255


for x in range(w):
    #print('...Pleas wait...',end="") #Druckt jede schleife 'Pleas Wait', so wei man das das Programm noch luft.
    for y in range(h):
        zx = 1.5 * (x - w / 2) / (0.5 * zoom * w) + moveX
        zy = 1.0 * (y - h / 2) / (0.5 * zoom * h) + moveY
        i = maxIter
        while zx * zx + zy * zy < 4 and i > 1:
            tmp = zx * zx - zy * zy + cX
            zy, zx = 2.0 * zx * zy + cY, tmp
            i -= 1
        # convert byte to RGB (3 bytes) RedGreenBlue
        pix[x, y] = (i << 21) + (i << 10) + i * 8

bitmap.show()
```
